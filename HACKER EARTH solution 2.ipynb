```python
test_mice5.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
train_mice5.drop('Employee_ID',axis=1, inplace=True)
test_mice5.drop('Employee_ID',axis=1, inplace=True)
```

```python
num_col = ['Education_Level','Time_of_service','Time_since_promotion','growth_rate','Travel_Rate','Post_Level','Pay_Scale','Work_Life_balance','VAR1','VAR2','VAR3','VAR4','VAR5','VAR6','VAR7']
```

```python
cat_columns = [col for col in train_mice5.columns if train_mice5[col].dtype =='O' ]
```

```python
encoder = LabelEncoder()

for col in cat_columns:
    train_mice5[col] = encoder.fit_transform(train_mice5[col])
    test_mice5[col] = encoder.transform(test_mice5[col])
```

```python
scaler = StandardScaler()

for col in num_col:
    train_mice5[col] = scaler.fit_transform(train_mice5[col].values.reshape(-1, 1))
    test_mice5[col] = scaler.transform(test_mice5[col].values.reshape(-1, 1))
```

```python
train_mice5.head()
```

```python
X = train_mice5.drop(['Attrition_rate'], axis=1)
y = train_mice5[['Attrition_rate']]
X_test = test_mice5
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=True)
```

```python
RF_reg=RandomForestRegressor(max_depth = 1, n_estimators= 307)
RF_reg.fit(X_train,y_train)
y_pred = RF_reg.predict(X_val)
rmse3 = np.sqrt(mean_squared_error(y_pred, y_val))
```

```python
rmse3
```

```python
RF_reg.fit(X, y)
predicts = RF_reg.predict(X_test)
```

```python
predicts
```

```python
xgb_reg = XGBRegressor(learning_rate=0.1, max_depth=1, min_child_weight=1, seed=0)

xgb_reg.fit(X_train, y_train)
y_pred = xgb_reg.predict(X_val)
rmse = np.sqrt(mean_squared_error(y_pred, y_val))
```

```python
xgb_reg.fit(X, y)
predict = xgb_reg.predict(X_test)
```
